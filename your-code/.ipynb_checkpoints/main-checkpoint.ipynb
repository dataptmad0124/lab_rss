### 7. Extract a list of entry titles.

### 8. Calculate the percentage of "Four short links" entry titles.

### 9. Create a Pandas data frame from the feed's entries.

```python
import pandas as pd
```

```python
entries=pd.DataFrame(pd.json_normalize(morningstar['entries']))
entries
```

### 10. Count the number of entries per author and sort them in descending order.

```python
entries.authors.value_counts()
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
entries['title length'] = [len(e) for e in title_list]

entries_sorted = entries.sort_values(by='title length', ascending=False)

entries_sorted[['title', 'author', 'title length']]

```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python
titles_with_machine_learning = entries_sorted[entries_sorted['summary'].str.contains('machine learning', case=False)]['title'].tolist()
```
